````python
import urllib.request as req
import bs4 as bs
# Step 0.
import pandas as pd

# Step1. 準備一個空的List
table = []
page = 1
while True:
    print("頁數:", page)
    url = "https://tabelog.com/tw/tokyo/rstLst/{}/?SrtT=rt".format(page)
    page = page + 1
    response = req.urlopen(url)
    html = bs.BeautifulSoup(response)

    r_list = html.find_all("div", {"class":"list-rst__body"})
    if len(r_list) == 0:
        break

    for r in r_list:
        en_name = r.find("a", {"class":"list-rst__rst-name-target"})
        en_name_href = en_name["href"]
        en_name_text = en_name.text
        area = r.find("div", {"class":"list-rst__area-genre"})
        area_text = area.text.strip()
        intro = r.find("p", {"class":"list-rst__pr-title"})
        if intro == None:
            intro_text = "無"
        else:
            intro_text = intro.text.strip()
        rating = r.find("span", {"class":"c-rating__val"})
        rating_text = rating.text
        prices = r.find_all("span", {"class":"c-rating-v3__val"})
        dinner_price = prices[0].text
        lunch_price = prices[1].text
        holiday = r.find("span", {"class":"list-rst__holiday-text"})
        holiday_text = holiday.text

        print(rating_text)
        print(en_name_text)
        print(en_name_href)
        print(area_text)
        print(intro_text)
        print(holiday_text)
        print("晚間價錢:", dinner_price, "午間價錢:", lunch_price)
        print("-" * 30)

        # Step2. 每一筆資料就是一個字典
        data = {
            "name(en)": en_name_text,
            "url": en_name_href,
            "area": area_text,
            "intro": intro_text,
            "rating": rating_text,
            "dinner_price": dinner_price,
            "lunch_price": lunch_price,
            "holiday": holiday_text
        }
        table.append(data)

# Step3. 轉成DataFrame儲存
df = pd.DataFrame(table,
          columns=["name(en)", "url", "area", "intro", "rating", "dinner_price", "lunch_price", "holiday"])
df.to_csv("tabelog_To.csv", encoding="utf-8")
df
````

````python
import urllib.request as req
import bs4 as bs
import pandas as pd

table = []

for i in range(60):
    print("第{}頁".format(i+1))
    url = "https://tabelog.com/tw/tokyo/rstLst/{}/?&SrtT=rt".format(i+1)
    response = req.urlopen(url)
    html = bs.BeautifulSoup(response)

    # find_all: 形態是個類似list的
    r_list = html.find_all("div", {"class":"list-rst__body"})
    # 一個餐廳都沒有的時候就是沒有下一頁了
    if len(r_list) == 0:
        break

    for r in r_list:
        en_name = r.find("a", {"class":"list-rst__rst-name-target"})
        en_comment = r.find("p", {"class":"list-rst__pr-title"})
        area = r.find("div", {"class":"cpy-area-genre"})
        rating = r.find("span", {"class":"c-rating__val"})
        holiday = r.find("span", {"class":"list-rst__holiday-text"})
        # 因為晚間和午間長得一樣
        prices = r.find_all("span", {"class":"c-rating-v3__val"})


        if en_comment == None:
            continue
        if holiday == None:
            continue
        if dinner_price == None:
            continue
        if lunch_price == None:
            continue

        en_name_href = en_name["href"]
        en_name_text = en_name.text
        area_text = area.text.strip()
        en_comment_text = en_comment.text.strip()
        rating_text = rating.text.strip()
        holiday_text = holiday.text
        dinner_price = prices[0].text.strip()
        lunch_price = prices[1].text.strip()

        print("店名:", en_name_text)
        print("地址:", area_text)
        print("評價:", en_comment_text)
        print("星等:", rating_text)
        print("晚間價位:", dinner_price)
        print("午間價位:", lunch_price)
        print("休市:", holiday_text)
        print("網址:", en_name_href)
        print("-" * 30)
        data = {
            "店名": en_name_text,
            "地址": area_text,
            "評價": en_comment_text,
            "星等": rating_text,
            "晚間價位": dinner_price,
            "午間價位": lunch_price,
            "休市": holiday_text,
            "網址": en_name_href
        }
        table.append(data)
df = pd.DataFrame(table, columns=["店名", "地址", "評價", "星等", "晚間價位", "午間價位", "休市", "網址"])
df.to_csv("tabelog_Tokyo.csv", encoding="utf-8")
df
````

# 抓住關鍵字

```python
1. 西方型: 英文 字母->詞彙->片語(phrase)
2. 東方型: 中文     字->詞彙

分詞: 我喜歡吃甜食 -> 我/喜歡/吃/甜食

哪個詞彙在整段話裡最重要:
1. 出現次數(TF): 高, 重要性越高
2. 常用程度(IDF): 低, 重要性越高

詞彙重要程度:
TF-IDF = 出現次數 * (1/常用程度)

!!! jieba函式庫: 預先統計了常用程度
DF = 詞彙出現在幾篇裡面/夠多篇文章
!!!      : 分詞
利用詞彙庫做出初步劃分->(沒在詞彙庫/混淆)上下文推測哪種分法最合理
```
