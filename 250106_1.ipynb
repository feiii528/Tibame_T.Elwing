#字典形態
```python
一副撲克牌: [排, 排, ...]
整班學生: [學生, 學生, ...]

字典型態: 把很多小東西聚集起來, 變成一個複雜的東西

字典在每一組的地方, 妳都要幫她準備一個key(是有形態)

key: 數字/字串
key-value: 連連看

名稱 = {
    "name":"Elwing",
    "height":175
}

? 如果你今天印出來: 發覺height在name前面有沒有問題
-> 沒問題, 字典沒有順序性(key=0,1,2...)

字典查詢: [key]
```

````python
student = {
    "name":"Elwing",
    "height":175
}
print(student)

print(student["height"])

# 口訣: 以前你怎麼使用名稱->字典[key]

# weight = 80
student["weight"] = 70
print(student)
# height = height + 5
student["height"] = student["height"] + 5
print(student)

# optional
for key in student:
    print("!", key, student[key])
````

# 標準表格格式
```python
CSV格式(Comma-Separated Values)

逗號/換行/雙引號
```

````python
import urllib.request as req
import json

while True:
    url = "https://www.cosmed.com.tw/api/getStore.aspx?t=store&c=&d=&s="
    f = req.urlopen(url)
    # content為bytes形態: 原始的01
    content = f.read()
    # bytes形態有一個叫做decode的專屬功能, 幫妳轉換形態str
    content = content.decode()
    # print(type(content))
    if "Invalid" in content:
        print("retry...")
    else:
        break
# print(type(content))
# 型態轉換: str -> dict/list
content = json.loads(content)
# print(type(content))
# content是字典，兩組: data, result
store_list = content["data"]
# store是一家店(字典)
for store in store_list:
    print(store["StoreNM"])
    print(store["Tel"])
    print(store["Address"])
    print("-" * 30)
````

````python
import urllib.request as req
import json

while True:
    url = "https://www.cosmed.com.tw/api/getStore.aspx?t=store&c=&d=&s="
    f = req.urlopen(url)
    content = f.read()
    content = content.decode()
    if "Invalid" in content:
        print("retry...")
    else:
        break

# 準備儲存的檔案
f = open("cosmed.csv", "w", encoding="utf-8")
f.write("名稱, 電話\n")
# 開啟檔案
content = json.loads(content)
store_list = content["data"]
for store in store_list:
    print(store["StoreNM"])
    print(store["Tel"])
    print(store["Address"])
    print("-" * 30)
    f.write(store["StoreNM"])
    f.write(",")
    f.write(store["Tel"])
    f.write("\n")

# 關閉
f.close()
````

````python
import json

url = "https://api.gamer.com.tw/anime/v1/danmu.php?videoSn=25562&geo=TW%2CHK&limit=50"
f = req.urlopen(url)
content = f.read()
content = content.decode()
content = json.loads(content)
danmu_list = content["data"]["danmu"]
# 準備儲存的檔案
f = open("danmu.csv", "w", encoding="utf-8")
f.write("時間, id, 彈幕")
for danmu in danmu_list:
    print(danmu["time"], ",",
       danmu["userid"], ",",
       file = f)

f.close()
````

````python
# pandas: 表格處理函式庫
import pandas as pd
import json

url = "https://api.gamer.com.tw/anime/v1/danmu.php?videoSn=25562&geo=TW%2CHK"
f = req.urlopen(url)

content = f.read()
content = content.decode()
content = json.loads(content)
danmu_list = content["data"]["danmu"]

# 幫妳把整個list轉換成pandas的表格形態(DataFrame)
df = pd.json_normalize(danmu_list)
# DataFrame有一個專屬功能叫做to_xxx
df.to_csv("danmu.csv", encoding="utf-8")
````
